```python
# 엑셀 파일이 있는 폴더 경로
folder_path = "C:/Users/munbe/Final project/서여의도점"

# 폴더 내의 모든 파일 목록 가져오기
file_list = os.listdir(folder_path)

# 엑셀 파일만 골라내기 (확장자가 .xlsx 또는 .xls인 파일)
excel_files = [file for file in file_list if file.endswith((".xlsx", ".xls"))]

# CSV 파일로 변환하여 저장할 폴더 경로
output_folder = "C:/Users/munbe/Final project/서여의도점"

# CSV 파일로 변환하고 저장
for excel_file in excel_files:
    # 엑셀 파일 경로
    excel_file_path = os.path.join(folder_path, excel_file)
    
    # CSV 파일 경로 (확장자 변경)
    csv_file_path = os.path.join(output_folder, excel_file.split('.')[0] + ".csv")
    
    # 엑셀 파일을 데이터프레임으로 읽어오기
    df = pd.read_excel(excel_file_path)
    
    # 데이터프레임을 CSV 파일로 저장 (인덱스는 저장하지 않음)
    df.to_csv(csv_file_path, index=False)

print("모든 엑셀 파일을 CSV 파일로 변환 및 저장 완료.")
```
